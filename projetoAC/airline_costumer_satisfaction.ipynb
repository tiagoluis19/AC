```python
from sklearn.neighbors import KNeighborsClassifier as KNN 
from sklearn.metrics import accuracy_score ,confusion_matrix

X_train = data.drop(['satisfaction'], axis=1)
y_train = data['satisfaction']
X_test = data_test.drop(['satisfaction'], axis=1)
y_test = data_test['satisfaction']

accuracies = []
f1_scores = []

# Loop through K values from 1 to 30
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)  # Create KNN with i neighbors
    knn.fit(X_train, y_train)  # Train the model
    y_pred = knn.predict(X_test)  # Predict
    acc = accuracy_score(y_true=y_test, y_pred=y_pred)  # Compute accuracy
    f1 = f1_score(y_true=y_test, y_pred=y_pred, average='weighted')  # Compute F1 score
    
    accuracies.append(acc)  # Store accuracy
    f1_scores.append(f1)  # Store F1 score

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(range(1, 21), accuracies, marker='o', linestyle='-', label='Accuracy')
plt.plot(range(1, 21), f1_scores, marker='s', linestyle='', label='F1 Score', color='red')

# Annotate accuracy and F1 score values at selected intervals
for i in range(1, 21, 3):  # Annotate every 3rd point for readability
    plt.text(i, accuracies[i-1], f'{accuracies[i-1]:.2f}', ha='center', va='bottom', fontsize=10)
    plt.text(i, f1_scores[i-1], f'{f1_scores[i-1]:.2f}', ha='center', va='top', fontsize=10, color='red')

plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Score")
plt.title("KNN Accuracy & F1 Score vs. Number of Neighbors (Standardized Data)")
plt.xticks(range(1, 21))
plt.grid(True)
plt.legend()
plt.show()
```
