```python
from sklearn.neighbors import KNeighborsClassifier as KNN 
from sklearn.metrics import accuracy_score ,confusion_matrix

X_train = data.drop(['satisfaction'], axis=1)
y_train = data['satisfaction']
X_test = data_test.drop(['satisfaction'], axis=1)
y_test = data_test['satisfaction']

accuracies = []

# Loop through K values from 1 to 20
for i in range(1, 21):
    knn = KNN(n_neighbors=i)  # Create KNN with i neighbors
    knn.fit(X_train, y_train)  # Train the model
    y_pred = knn.predict(X_test)  # Predict
    acc = accuracy_score(y_true=y_test, y_pred=y_pred)  # Compute accuracy
    accuracies.append(acc)  # Store accuracy

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 21), accuracies, marker='o', linestyle='-')
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy Score")
plt.title("KNN Accuracy vs. Number of Neighbors")
plt.xticks(range(1, 21))
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Assuming 'data' and 'data_test' are already defined

X_train = data.drop(['satisfaction'], axis=1)
y_train = data['satisfaction']
X_test = data_test.drop(['satisfaction'], axis=1)
y_test = data_test['satisfaction']

# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

accuracies = []

# Loop through K values from 1 to 20
for i in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors=i)  # Create KNN with i neighbors
    knn.fit(X_train, y_train)  # Train the model
    y_pred = knn.predict(X_test)  # Predict
    acc = accuracy_score(y_true=y_test, y_pred=y_pred)  # Compute accuracy
    accuracies.append(acc)  # Store accuracy

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 31), accuracies, marker='o', linestyle='-')
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy Score")
plt.title("KNN Accuracy vs. Number of Neighbors (Standardized Data)")
plt.xticks(range(1, 31))
plt.grid(True)

plt.show()
```
